Task 1: Load and Explore the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
data = pd.read_csv("data.csv")
```

Task 2: Basic Data Analysis

```python
# Display the first 5 rows
print(data.head())
```

```python
# Check the data types of columns
print(data.info())

# Check for missing values
print(data.isnull().sum())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
# Compute statistics
print(data.describe())
```

Task 3: Data Visualization

```python
# Group by 'city' and compute the mean of the 'price' column
group_means = data.groupby('city')['price'].mean()
print(group_means)
```

```python
# Example: Inspect if certain groups have significantly different means
print(group_means.sort_values(ascending=False))
```

```python
# Average numerical values per category
group_means.plot(kind='bar', title='Average Numerical Values per Category', ylabel='Average Value', xlabel='Category')
plt.show()
```

```python
 # Plotting histogram for the 'price' column (you can replace it with any other numerical column)
data['price'].plot(kind='hist', bins=20, title='Distribution of Price', xlabel='Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot between 'price' and 'sqft_living'
sns.scatterplot(x='sqft_living', y='price', data=data)
plt.title('Scatter Plot: sqft_living vs Price')
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.show()
```
